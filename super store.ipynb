# BAR CHART

```python
# Sales analysis based on Region using Bar Chart

superstore.groupby("Region")["Sales"].sum().plot.bar()
```

```python
# Profit analysis based on Region using Bar Chart

superstore.groupby("Region")["Sales"].sum().plot.bar()
```

```python
# Sales Analysis based on segment using Bar Chart

superstore.groupby("Customer Segment")["Sales"].sum().plot.bar()
```

```python
# Profit Analysis based on Segment using Bar Chart

superstore.groupby("Customer Segment")["Profit"].sum().plot.bar()
```

```python
# Sales analysis based on Category of the product using Bar Chart

superstore.groupby("Product Category")["Sales"].sum().plot.bar()
```

```python
# Profit Analysis based on Category of the product using Bar Chart

superstore.groupby("Product Category")["Profit"].sum().plot.bar()
```

```python
# Sales Analysis based on State using Bar Chart

superstore.groupby("State or Province")["Sales"].sum().plot.bar()
```

```python
# Profit Analysis based on State using Bar Chart

superstore.groupby("State or Province")["Profit"].sum().plot.bar()
```

```python
# Profit Analysis based on ID of the customer using Bar Chart

superstore.groupby("Customer ID")["Profit"].sum().plot.bar()
```

```python
# Profit Analysis based on Shipping Cost using Bar Chart

superstore.groupby("Shipping Cost")["Profit"].sum().plot.bar()
```

```python
# Profit Analysis based on Quantity ordered using Bar Chart

superstore.groupby("Quantity ordered new")["Profit"].sum().plot.bar()
```

```python
# Sales Analysis based on City using Bar Chart

superstore.groupby("City")["Sales"].sum().plot.bar()
```

```python
# Profit Analysis based on City using Bar Chart

superstore.groupby("City")["Profit"].sum().plot.bar()
```
